```python
import matplotlib.pyplot as plt
import pandas as pd

# Data for the visualization
data = {
    'Year': [2000, 2005, 2010, 2015, 2020, 2022, 2024, 2025],
    'USD_Share': [71.5, 66.5, 61.8, 65.7, 58.9, 58.4, 57.8, 56.3],
    'Gold_Share': [13.0, 10.5, 11.0, 9.0, 14.0, 15.5, 17.6, 19.5] # 2025 estimated based on recent surge
}

df = pd.DataFrame(data)

# Create the plot
fig, ax1 = plt.subplots(figsize=(12, 7))

# Plot USD Share on the left Y-axis (Blue)
color = 'tab:blue'
ax1.set_xlabel('Year', fontsize=12, fontweight='bold')
ax1.set_ylabel('USD Share of Global Reserves (%)', color=color, fontsize=12, fontweight='bold')
ax1.plot(df['Year'], df['USD_Share'], color=color, marker='o', linewidth=3, label='USD Share')
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(True, linestyle='--', alpha=0.5)

# Create a second Y-axis for Gold Share (Gold color)
ax2 = ax1.twinx()
color = '#D4AF37' # Gold color
ax2.set_ylabel('Gold Share of Global Reserves (%)', color=color, fontsize=12, fontweight='bold')
ax2.plot(df['Year'], df['Gold_Share'], color=color, marker='s', linewidth=3, linestyle='-', label='Gold Share')
ax2.tick_params(axis='y', labelcolor=color)

# Add annotations for key events
plt.title('The Great Divergence: De-Dollarization Trend (2000–2025)', fontsize=16, fontweight='bold', pad=20)
ax1.annotate('Sanctions Trigger\nAccelerated Gold Buying', xy=(2022, 58.4), xytext=(2018, 50),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=10, bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="black", lw=1))

ax2.annotate('Gold "Crossover"\nReserves Surge', xy=(2025, 19.5), xytext=(2021, 22),
             arrowprops=dict(facecolor='#D4AF37', shrink=0.05), fontsize=10, color='#997f00', fontweight='bold')

# Show the plot
fig.tight_layout()
plt.show()
```
