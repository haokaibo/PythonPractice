```python
from typing import Optional
from collections import deque

# Definition for singly-linked list.
class ListNode:
     def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class QueueByDeque:
    def __init__(self):
        self.queue = deque()

    def push(self, item):
        self.queue.append(item)

    def pop(self):
        if len(self.queue) <= 0:
            return None
        else:
            return self.queue.popleft()

    def peek(self):
        if len(self.queue) <= 0:
            return None
        else:
            return self.queue[0]

    def last(self):
        if len(self.queue) <= 0:
            return None
        else:
            return self.queue[-1]

    def empty(self):
        return len(self.queue) == 0

    def __str__(self):
        return str(self.queue)


class Solution:
    def iterateListNode(self, listNode, queue) -> QueueByDeque:
        current = listNode
        queue.push(current.val)
        while(current.next is not None):
            current = current.next
            queue.push(current.val)          


    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        q1 = QueueByDeque()
        q2 = QueueByDeque()

        self.iterateListNode(l1, q1)
        self.iterateListNode(l2, q2)
        
        last = None
        previousSum = 0
        while(not q1.empty() and not q2.empty()):
            currentSum = 0 if q1.empty() else q1.pop() 
            + 0 if q2.empty() else q2.pop() + previousSum // 10
            current = ListNode(currentSum % 10, last)
            last = current
            previousSum = currentSum

        return last
     
```

```python
num1 = [2, 4, 3]
num2 = [5, 6, 4]

def createNodeList(num):
    previousNode = None
    for i in num:
        head = ListNode(i, previousNode)
        previousNode = head
    return head

def printNodeList(nodeList):
    finalStr = str(nodeList.val)
    current = nodeList.next
    while(current is not None):
        finalStr += f'-> {current.val}'
        current = current.next
        
l1 = createNodeList(num1)
l2 = createNodeList(num2)

for l in [l1, l2]:
    printNodeList(l1)
```
