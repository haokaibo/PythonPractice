You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order, and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.

 

Example 1:


Input: l1 = [2,4,3], l2 = [5,6,4]
Output: [7,0,8]
Explanation: 342 + 465 = 807.
Example 2:

Input: l1 = [0], l2 = [0]
Output: [0]
Example 3:

Input: l1 = [9,9,9,9,9,9,9], l2 = [9,9,9,9]
Output: [8,9,9,9,0,0,0,1]
 

Constraints:

The number of nodes in each linked list is in the range [1, 100].
0 <= Node.val <= 9
It is guaranteed that the list represents a number that does not have leading zeros.


```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

# 2 -> 4 -> 3 
# 5 -> 6 -> 4
# 7 -> 0 -> 8

# 2 + 5 = 7
# 4 + 6 = 10 （10 // 10 = 1， 10 % 10  = 0）
# 3 + 4 + 1 = 8

# 1. Loop through the ListNodes and 
# 2. Do the calculation
    # 2.1 handle the carrying over when reaching ten
# 3. Create a new node list to hold the result

# 1
# 9 -> 3

# 0 -> 4

class Solution:

    def addNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode], carrying_over: int =0) -> Optional[ListNode]:
        if l1 is None and l2 is None and carrying_over == 0:
            return None

        l1_val = 0 if l1 is None else l1.val
        l2_val = 0 if l2 is None else l2.val
        sum_of_nums = l1_val + l2_val + carrying_over

        next_l1 = None if l1 is None else l1.next
        next_l2 = None if l2 is None else l2.next

        result = ListNode(sum_of_nums % 10, self.addNumbers(next_l1, next_l2, sum_of_nums//10))
        return result

    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:

        return self.addNumbers(l1, l2)

```

```python
num1 = [2, 4, 3]
num2 = [5, 6, 4]

def createNodeList(num):
    previousNode = None
    for i in num[::-1]:
        head = ListNode(i, previousNode)
        previousNode = head
    return head

def printNodeList(nodeList):
    finalStr = str(nodeList.val)
    current = nodeList.next
    while(current is not None):
        finalStr += f'-> {current.val}'
        current = current.next
    print(finalStr)
        
l1 = createNodeList(num1)
l2 = createNodeList(num2)

for l in [l1, l2]:
    printNodeList(l)
```

```python
result = Solution().addTwoNumbers(l1, l2)
printNodeList(result)
```
